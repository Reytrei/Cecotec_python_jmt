# Prueba Python

##### Importar librerias necesarias

```python
import requests
import pandas as pd
import numpy as np
```

##### Descargar el archivo y abrirlo

```python
url = "https://cecobi.cecotec.cloud/ws/getstocksabanax.php"
output = "StockCecotecNacional.csv"

response = requests.get(url)
with open(output, 'wb') as f:
    f.write(response.content)
```

##### Importar el dataset e imprimir la cabecera

```python
stock = pd.read_csv(output, sep= ";", encoding = 'ISO-8859-1')
stock.head()
```

##### Explorar valores nulos

```python
stock.info()
```

##### Filtrar Referencias que empiezan con "V" y exportar el dataset

```python
stock_v = stock[stock['Referencia'].str.startswith('V')]
print(stock_v.shape)
stock_v.to_csv('stock_v.csv', encoding= 'ISO-8859-1')
stock_v.head()
```

##### Filtrar Referencias que empiezan con "P" y exportar dataset

```python
stock_p = stock[stock['Referencia'].str.startswith('P')]
print(stock_p.shape)
stock_p.to_csv('stock_p.csv', encoding= 'ISO-8859-1')
stock_p.head()
```

##### Filtrar Referencias que empiezan con un numero

```python
stock_num = stock[stock['Referencia'].str.match('^\d', na=False)]
print(stock_num.shape)
stock_num.head()
```

##### Filtrar Restantes que no cumplen con ninguno de los requisitos anteriores y exportar dataset

```python
stock_resto = df_rest = stock[~(stock['Referencia'].str.match('^V', na=False) |
                  stock['Referencia'].str.match('^P', na=False) |
                  stock['Referencia'].str.match('^\d', na=False))]
print(stock_resto.shape)
stock_resto.to_csv('stock_resto.csv', encoding= 'ISO-8859-1')
stock_resto.head()
```

##### Crear dataset oportunidad de ventas y exportarlo
Se filtra valores de stock fisico mayores a 0

```python
oportunidad_venta = stock_num[stock_num['Stock Fisico'] > 0]
print(oportunidad_venta.shape)
oportunidad_venta.to_csv('oportunidad_venta.csv', encoding= 'ISO-8859-1')
oportunidad_venta.head()
```

##### Crear columna "Futura"
Futura es la suma de las columnas Mar, Puerto y Despachado

```python
stock_num['Futura'] = stock_num['Mar'] + stock_num['Puerto'] + stock_num['Despachado']
print(stock_num.shape)
stock_num.head()
```

##### Filtrar valores para dataset y exportar futura comercial

Se filtran valores de stock comercial y futura iguales a 0

```python
futura_comercial_0 = stock_num[(stock_num['Stock Comercial'] == 0) & (stock_num['Futura'] ==0)]
print(futura_comercial_0.shape)
futura_comercial_0.to_csv('futura_comercial_0.csv', encoding= 'ISO-8859-1')
futura_comercial_0.head()
```

##### Crear columna Remanente
La columna remanente se crea por la diferencia de stock fisico y stock comercial

```python
stock_num['Remanente'] = stock_num['Stock Fisico'] - stock_num['Stock Comercial']
stock_num.head()
```

##### Crear columna porcentaje y exportar dataset final

Se divide la columna Remanente sobre Stock Fisico. En el caso  de que Stock Fisico sea 0 se le asigna un 0 para no generar valores nulos. Luego se exporta dataset

```python
stock_num['Porcentaje'] = np.where(stock_num['Stock Fisico'] == 0, 0, stock_num['Remanente'] / stock_num['Stock Fisico'])
stock_num['Porcentaje'] = stock_num['Porcentaje'] * 100
stock_num.to_csv('stock_num.csv', encoding= 'ISO-8859-1')
stock_num.head()
```
