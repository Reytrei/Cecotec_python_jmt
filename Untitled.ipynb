```python
import requests
import pandas as pd
import numpy as np
```

```python
url = "https://cecobi.cecotec.cloud/ws/getstocksabanax.php"
output = "StockCecotecNacional.csv"

response = requests.get(url)
with open(output, 'wb') as f:
    f.write(response.content)
```

```python
stock = pd.read_csv(output, sep= ";", encoding = 'ISO-8859-1')
stock.head()
```

```python
stock.info()
```

```python
stock_v = stock[stock['Referencia'].str.startswith('V')]
print(stock_v.shape)
stock_v.head()
```

```python
stock_l = stock[stock['Referencia'].str.startswith('L')]
print(stock_l.shape)
stock_l
```

```python
stock_num = stock[~stock['Referencia'].str.startswith('V', 'L')]
print(stock_num.shape)
stock_num.head()
```

```python
oportunidad_venta = stock_num[stock_num['Stock Fisico'] > 0]
print(oportunidad_venta.shape)
oportunidad_venta.head()
```

```python
stock_num['Futura'] = stock_num['Mar'] + stock_num['Puerto'] + stock_num['Despachado']
print(stock_num.shape)
stock_num.head()
```

```python
futura_comercial_0 = stock_num[(stock_num['Stock Comercial'] == 0) & (stock_num['Futura'] ==0)]
print(futura_comercial_0.shape)
futura_comercial_0.head()
```

```python
stock_num['Remanente'] = stock_num['Stock Fisico'] - stock_num['Stock Comercial']
stock_num.head()
```

```python
stock_num['Porcentaje'] = np.where(stock_num['Stock Fisico'] == 0, 0, stock_num['Remanente'] / stock_num['Stock Fisico'])
stock_num['Porcentaje'] = stock_num['Porcentaje'] * 100
stock_num.head()
```
